# optimal linear regression
In this notebook, I am going to implement `backward elimination` concept to find the most powerful predictable features and get the best result.

```python
#import libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
import statsmodels.api as sm
```

```python
dataset = pd.read_csv("Data.csv")
```

```python
dataset.head(1)
```

```python
#import dataset
dataset = dataset.drop(['Unnamed: 0', 'Name of show'], axis = 1)
```

```python
dataset = dataset.dropna()
```

```python
X = dataset.iloc[:, [1, 2, 3, 6, 9, 11, 12, 14, 15, 17]]
y = dataset.iloc[:, 16].values
```

```python
X = pd.get_dummies(X, prefix_sep='_', drop_first=True)
```

```python
X.head(1)
```

```python
X.shape
```

```python
X = np.append(arr = np.ones((X.shape[0], 1)).astype(int), values = X, axis = 1)
X_opt = X[:, 0:(X.shape[1])]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```
